```python
import numpy as np
import math
x=np.random.uniform(-10,10)
eta=0.1
iter_max=50
iter=1
while iter<=iter_max:
  slope=4*x*x*x+6*x
  step_length=-eta*slope
  x=x+step_length
  iter=iter+1
obj=pow(x,4)+3*pow(x,2)+10
print("optimal solution=",obj,"at X=",x)
```
