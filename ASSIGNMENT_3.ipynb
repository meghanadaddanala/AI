```python
import numpy as np
import math
x=np.random.uniform(-10,10)
y=np.random.uniform(-10,10)
eta=0.1
iter_max=100
iter=1
while iter<=iter_max:
  grad_x=6*x
  grad_y=-5*(math.exp(-y))
  del_x=-eta*grad_x
  del_y=-eta*grad_y
  x=x+del_x
  y=y+del_y
  iter=iter+1
obj=3*x*x+5*math.exp(-y)+10
print('optimal solution=',obj,'at X=',x,'at Y=',y)
```
